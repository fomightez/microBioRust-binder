### Demonstrate Genbank to FASTA

Let's set this up by putting a Genbank file in our working directory.

```python
!cp ../microbiorust-py/test_input.gbk .
```

Verify it is here in our working directory:

```python
ls
```

```python
from microbiorust import gbk_to_faa
result = gbk_to_faa("test_input.gbk")
for r in result:
    print(r)
```

Let's add printing each to a file:

```python
from microbiorust import gbk_to_faa
result = gbk_to_faa("test_input.gbk")
for i,r in enumerate(result):
    %store r >fasta_{i}.fa
    print(r)
```

Let's look at the content of one of those generated files:

```python
from pathlib import Path
print(Path("fasta_1.fa").read_text())
```
